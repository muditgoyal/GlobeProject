## Univariate Feature Selection

### Useful functions

```python
def explore_bin_averages(df,target):
    colormap = plt.cm.viridis
    
    for col in df.drop(target,axis=1).columns:
        avg_dict = {}
        for value in df[col].unique():
            avg_dict[value] = np.mean(df[target][df[col]==value])
        avg_df = pd.DataFrame.from_dict(avg_dict,orient='index')
        avg_df.plot.bar(colormap=colormap,title=col)
        
def distcorr(X, Y):
    """ Compute the distance correlation function
    
    >>> a = [1,2,3,4,5]
    >>> b = np.array([1,2,9,4,4])
    >>> distcorr(a, b)
    0.762676242417
    """
    X = np.atleast_1d(X)
    Y = np.atleast_1d(Y)
    if np.prod(X.shape) == len(X):
        X = X[:, None]
    if np.prod(Y.shape) == len(Y):
        Y = Y[:, None]
    X = np.atleast_2d(X)
    Y = np.atleast_2d(Y)
    n = X.shape[0]
    if Y.shape[0] != X.shape[0]:
        raise ValueError('Number of samples must match')
    a = squareform(pdist(X))
    b = squareform(pdist(Y))
    A = a - a.mean(axis=0)[None, :] - a.mean(axis=1)[:, None] + a.mean()
    B = b - b.mean(axis=0)[None, :] - b.mean(axis=1)[:, None] + b.mean()
    
    dcov2_xy = (A * B).sum()/float(n * n)
    dcov2_xx = (A * A).sum()/float(n * n)
    dcov2_yy = (B * B).sum()/float(n * n)
    dcor = np.sqrt(dcov2_xy)/np.sqrt(np.sqrt(dcov2_xx) * np.sqrt(dcov2_yy))
    return dcor
```

### Pearson Correlations

### Distance Correlation

```python
dist_cor_dict = {}

for col in df.drop('Valuation Increase',axis=1).columns:    
    dist_cor_dict[col] = distcorr(df['Valuation Increase'],df[col])

dcor_df = pd.DataFrame.from_dict(dist_cor_dict,orient='index')
dcor_df.rename_axis({0:'Distance Correlation'},axis=1,inplace=True)
dcor_df.sort_values(['Distance Correlation'],ascending = True,inplace=True)
dcor_df
```

```python
dcor_df.plot.barh(colormap=colormap,title='Distance Correlation to Valuation Increase')
```

## Random Forest or Impurity Minimizing Selection

```python
modl = RFR(n_estimators = 1000, n_jobs=-1, criterion='mse')

modl.fit(df.drop('Valuation Increase', axis=1), df['Valuation Increase'])

print(modl.score(df.drop('Valuation Increase', axis=1), df['Valuation Increase']))

# Look at importance of features for random forest
imp = pd.DataFrame()
imp['Importance']=modl.feature_importances_
imp.set_index(df.drop('Valuation Increase',axis=1).columns,inplace=True)
imp.sort_values([ 'Importance' ],ascending = True,inplace=True)
```

```python
imp.plot.barh(colormap=colormap)
```

```python
modl = RFR(n_estimators = 1000, n_jobs=-1, criterion='mae')

modl.fit(df.drop('Valuation Increase', axis=1), df['Valuation Increase'])

print(modl.score(df.drop('Valuation Increase', axis=1), df['Valuation Increase']))

# Look at importance of features for random forest
imp = pd.DataFrame()
imp['Importance']=modl.feature_importances_
imp.set_index(df.drop('Valuation Increase',axis=1).columns,inplace=True)
imp.sort_values([ 'Importance' ],ascending = True,inplace=True)
```

```python
imp.plot.barh(colormap=colormap)
```

## Lasso Stability Selection

```python
#Data gets scaled automatically by sklearn's implementation
 
rlasso = RandomizedLasso(alpha=0.025)
rlasso.fit(df.drop('Valuation Increase',axis=1), df['Valuation Increase'])
 
# print("Features sorted by their score:")
# print(sorted(zip(map(lambda x: round(x, 4), rlasso.scores_), 
#                  df.drop('Valuation Increase',axis=1).columns), reverse=True))



ss_df = pd.DataFrame()
ss_df['lasso score']=rlasso.scores_
ss_df.set_index(df.drop('Valuation Increase',axis=1).columns,inplace=True)
ss_df.sort_values(['lasso score'],ascending = True,inplace=True)
```
